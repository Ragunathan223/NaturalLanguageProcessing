```python
# 2. Print all possible definitions of the word ‘old’ and their corresponding antonym.
```

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("old"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(set(synonyms))
print("---------------------------------")
print('The corresponding antonyms for the word old is ',set(antonyms))
```

```python
# 3.Print hypernyms of the list of all animals/birds in the story
```

```python
animals = ['lion', 'hippo', 'rhino', 'gazelle', 'monkey','chicken', 'ostrich', 'vulture', 'wolf', 'hyena']
for i in animals:
    My_sysn = wordnet.synsets(i)[0]
    print(f'The Hypernym for the word {i} is:',My_sysn.hypernyms(),'\n')
```

```python
# 4. Find the animal/bird from the list that is most similar to ‘hyena’ using Wu & Palmer word similarity.
```

```python
from nltk.corpus import wordnet
for i in animals: 
    syn1 = wordnet.synsets(i)[0]
    syn2 = wordnet.synsets('hyena')[0]
    sim = syn1.wup_similarity(syn2)
    print(i,':',sim)
```

```python
# 5. Find all the wrongly spelled words and print them along with their correct spellings. 
#    NOTE: the wrongly spelled words appear in red color in the story for reference.
```

```python
# 6. Using FreqDist, Find frequency of a new list of words without stopwords. 
#   NOTE: Retain the original frequency of occurrence of each word.
```

```python
import nltk
from nltk.corpus import stopwords
stop_words = (stopwords.words('english'))
```

```python
filtered_words = []

for w in words:
    if w.lower() not in stopwords.words('english'):
        filtered_words.append(w)
print(filtered_words)
```

```python
from nltk import FreqDist
fd = FreqDist(filtered_words)
fd
```

```python
# 7. From the new list, create a final list of words whose frequency is greater than 3 and plot the frequency distribution 
#    graph of the final list with a title “Words with frequency greater than 3”.
```

```python
final_words = [word for word in filtered_words if fd[word] > 3]
freq_dist_final = FreqDist(final_words)
freq_dist_final.plot(title="Words with frequency greater than 3")
```
